```python
import pandas as pd
import numpy as np
```

```python
training_data = pd.read_csv('/Users/parikshitnarang/Desktop/twitter_train.csv')
```

```python
training_data
```

```python
testing_data = pd.read_csv('/Users/parikshitnarang/Desktop/twitter_test.csv')
```

```python
testing_data.text[0]
```

```python
x_train = np.array(training_data['text'])
x_train
```

```python
x_test = np.array(testing_data['text'])
x_test
```

```python
y_train = np.array(training_data['airline_sentiment'])
y_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
count_vec = TfidfVectorizer(stop_words='english', max_features=2300, ngram_range=(1,4))
```

```python
x_train_features = count_vec.fit_transform(x_train)
x_train_features
```

```python
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features = count_vec.transform(x_test)
```

```python
x_test_features
```

```python
x_test_features.todense()
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(C=2.5)
```

```python
svc.fit(x_train_features, y_train)
```

```python
y_pred = svc.predict(x_test_features)
```

```python
y_pred
```

```python
np.savetxt('/Users/parikshitnarang/Desktop/pred4.csv', y_pred, delimiter=',', fmt='%s')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=150, max_depth=None)
```

```python
rf.fit(x_train_features, y_train)
```

```python
y_pred = rf.predict(x_test_features)
```

```python
np.savetxt('/Users/parikshitnarang/Desktop/pred4.csv', y_pred, delimiter=',', fmt='%s')
```
